```python
#Switching string with ints

sex = {'Male': 0, 'Female': 1}

df.Gender = [sex[item] for item in df.Gender]

spending = {'Low': 2, 'Average': 4, 'High': 6}

df.Spending_Score = [spending[item] for item in df.Spending_Score]
```

```python
df_min_max_scaled = df.copy()
  

for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  

print(df_min_max_scaled)
```

```python
def testTrainSplit(data):
  length = len(df_min_max_scaled.index)
  splitLimit = int(length * 0.7)
  train = data[0: splitLimit]
  test = data[splitLimit + 1: length]
  return train, test
```

```python
train, test = testTrainSplit(df_min_max_scaled)
```
