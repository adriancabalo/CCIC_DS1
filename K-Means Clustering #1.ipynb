```python
df_min_max_scaled = df.copy()
  

for column in df_min_max_scaled.columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())    
  

print(df_min_max_scaled)
```

```python
def testTrainSplit(data):
  length = len(df_min_max_scaled.index)
  splitLimit = int(length * 0.7)
  train = data[0: splitLimit]
  test = data[splitLimit + 1: length]
  return train, test
```

```python
from sklearn.cluster import KMeans
ks=[]
for i in range(1,5):
    ks.append(i)
df = df.copy([['Spending_Score', 'State_Purchased','Family_Size','Age','Gender']])
```

```python
wcss = []


for i in range(1, 5):
    clustering = KMeans(n_clusters=i, init='k-means++', random_state=42)
    clustering.fit(df)
    wcss.append(clustering.inertia_)
    
data = {"KValue":ks,"Variation":wcss}
df2 = pd.DataFrame(data)
sns.lineplot(data=df2, x = "KValue", y = "Variation");
```
