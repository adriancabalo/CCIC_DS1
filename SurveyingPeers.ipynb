```python
school = [["\nGrandview\n", "*", "*", "*", "*"],
          ["\nSmoky Hill\n",  "*", "*", "*", "*"],
          ["\nEaglecrest\n", "*"],
          ["\nCherokee Trail\n", "*", "*", "*" ],
          ["\nOverland\n",  "*"],
          ["\nMs.Spindt\n",  "*"]]

drive = [["\nYes\n", "*", "*", "*", "*", "*", "*", "*",],
         ["\nNo\n", "*", "*", "*", "*", "*", "*", "*",]]

homeSchool = [["\nBus\n", "*", "*", "*", "*", "*", "*", "*", "*",],
              ["\nDropped off\n", "*", "*",],
             ["\nI Drive\n", "*", "*", "*",],
             ["\nMs.Spindt\n",  "*",]]

siblings = [["\nYes, more than one\n", "*","*","*","*","*","*","*","*","*","*"],
            ["\nYes, just one\n","*","*"],
            ["\nNo\n","*","*","*"]]

grade = [["9th","*"],
         ["10th","*","*"],
         ["11th","*","*","*","*"],
         ["12th","*","*","*","*","*","*"]]

print("<What school do you go to?> ")
for i in school:
    for x in i:
        print(x, end = " ")

print("\n\n<Do you drive?> ")
for i in drive:
    for x in i:
        print(x, end = " ")
print("\n\n<How do you get to your homeschool?> ")
for i in homeSchool:
    for x in i:
        print(x, end = " ")
        
print("\n\n<Do you have any siblings?> ")
for i in siblings:
    for x in i:
        print(x, end = " ")
        
print("\n\n<What grade are you in?>\n ")        
for i in grade:
    for x in i:
        if(x == "*"):
            print(x, end = " ")
        elif(x == "9th"):
            print("9th")
        elif(x == "10th"):
            print("\n10th")
        elif(x == "11th"):
            print("\n11th")
        elif(x == "12th"):
            print("\n12th")
            


            
# Although half of the class is able to drive, a majority still takes the bus to their homeschools
       
```
