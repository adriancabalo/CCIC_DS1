```python
def movieList():
    
    pageNum = 1
    movies = []
    
    while True:
        url = 'https://api.themoviedb.org/3/movie/{movie_id}/credits?api_key=87983fbe976988a9e22330ebf030a1ae&page=1' + str(pageNum)
        req = requests.get(url)
        if not req.ok:
            break
        movies = movies +req.json()["results"]
        pageNum += 1
        
    return movies
```

```python
listOfMovies = movieList()
```
