```python
def movieList():
    
    pageNum = 1
    movies = []
    while True:
        url = 'https://api.themoviedb.org/3/movie/?api_key=87983fbe976988a9e22330ebf030a1ae&sort_by=revenue.desc&with_genres=28&page=' + str(pageNum)
        req = requests.get(url)
        if not req.ok:
            break
        movies = movies + req.json()['results']
        pageNum += 1
        
    return movies
```

```python
listOfMovies = movieList()
```

```python
idList = []

for i in listOfMovies:
        ident = i['id']
        idList.append(ident)

print(idList)
```

```python
columns = ['film', 'revenue']
df = pd.DataFrame(columns=columns)
```

```python
for film in listOfMovies:
    film_revenue = requests.get('https://api.themoviedb.org/3/movie/'+ str(film['id']) +'?api_key='+ '87983fbe976988a9e22330ebf030a1ae' +'&language=en-US')
    film_revenue = film_revenue.json()

    df.loc[len(df)]=[film['title'],film_revenue['revenue']] 
```
